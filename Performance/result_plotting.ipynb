```python
class plot_result():
    def plot_acc_graph(self, lstm_test, lstm_train, maxpool_test, maxpool_train, attn_test, attn_train, pat_score):
        #plt.gca().cla()
        plt.subplots(figsize=(15,8))
        #plt.figure(1, figsize=(15, 9))
        plt.title('Test Accuracies Based on Patients ' + pat_score + ' Experiment Values')
        plt.plot(lstm_test, color='green', label='JK-LSTM Network-Testing', linewidth=2)
        plt.plot(lstm_train, color='c', label='JK-LSTM Network-Training', linewidth=2)
        
        plt.plot(maxpool_test, color='red', label='JK-MAXPOOL Network-Testing', linewidth=2)
        plt.plot(maxpool_train, color='orange', label='JK-MAXPOOL Network-Training', linewidth=2)
        
        plt.plot(attn_test, color='navy', label='JK-Attn Network-Testing', linewidth=2)
        plt.plot(attn_train, color='magenta', label='JK-Attn Network-Training', linewidth=2)
        
        plt.xlabel('Iterations')
        plt.ylabel('Accuracies')
        plt.legend()
        #img1_name = "loss_" + datetime.datetime.now().strftime("%I%M%p%B%d%Y") + ".png"
        #plt.show()
        plt.savefig("accuracy_graph_" + pat_score + ".png")
```

```python
plt_obj = plot_result()
plt_obj.plot_acc_graph(moca_lstm_test_acc, moca_lstm_train_acc, moca_maxpool_test_acc, moca_maxpool_train_acc, 
                       moca_attention_test_acc, moca_attention_train_acc, "MOCA")
```
